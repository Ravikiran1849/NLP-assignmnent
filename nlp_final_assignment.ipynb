**Sample Dataset**

```python
tweets = [
    "I love playing cricket!",
    "My dad is the greatest person on the world.",
    "My friends are the backdone of my life.",
    "Text processing includes tokenization and stemming.",
    "SpaCy and NLTK are popular NLP libraries."
]
```

**Tokenization and stemming and lemmatization using NLTK:**

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from collections import Counter
import matplotlib.pyplot as plt

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt_tab')


stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

nltk_tokens = []
nltk_stems = []
nltk_lemmas = []

for tweet in tweets:
    tokens = word_tokenize(tweet.lower())
    nltk_tokens.extend(tokens)
    nltk_stems.extend([stemmer.stem(t) for t in tokens])
    nltk_lemmas.extend([lemmatizer.lemmatize(t) for t in tokens])
```

**Tokenization and Lemmatization using SpaCy**

```python
import spacy
nlp = spacy.load("en_core_web_sm")

spacy_tokens = []
spacy_lemmas = []

for tweet in tweets:
    doc = nlp(tweet.lower())
    spacy_tokens.append([token.text for token in doc])
    spacy_lemmas.append([token.lemma_ for token in doc])
```

**Visualisation for Token frequency and comparsion**

```python
from collections import Counter
import matplotlib.pyplot as plt

def plot_token_freq(data, title):
    flat = [item for sublist in data for item in sublist]
    counts = Counter(flat)
    common = counts.most_common(10)
    words, freq = zip(*common)

    plt.figure(figsize=(8,4))
    plt.bar(words, freq, color='skyblue')
    plt.title(title)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

plot_token_freq(nltk_tokens, "NLTK Tokens")
plot_token_freq(nltk_stems, "NLTK Stems")
plot_token_freq(nltk_lemmas, "NLTK Lemmas")
plot_token_freq(spacy_tokens, "SpaCy Tokens")
plot_token_freq(spacy_lemmas, "SpaCy Lemmas")
```

**2.Load and Preprocess Data**

```python
import pandas as pd
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')

try:
    df = pd.read_csv('emotion.csv')
    df = df[['text', 'emotion']]
except FileNotFoundError:
    df = pd.DataFrame({'text': [], 'emotion': []})


stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_text(text):

    if not isinstance(text, str):
        return ""

    text = re.sub(r"http\S+|[^a-zA-Z\s]", "", text)
    tokens = nltk.word_tokenize(text.lower())
    tokens = [lemmatizer.lemmatize(w) for w in tokens if w not in stop_words]
    return ' '.join(tokens)

if not df.empty:
    df['clean_text'] = df['text'].apply(clean_text)
```

**3.Train Logistic Regression Model**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Load the dataset from CSV file
df = pd.read_csv("/content/emotion-labels-test.csv")  # Make sure the file is in the same directory
df.rename(columns={"text": "clean_text", "label": "emotion"}, inplace=True)  # Match column names
X_train, X_test, y_train, y_test = train_test_split(df['clean_text'], df['emotion'], test_size=0.2, random_state=42)

vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train_vec, y_train)
y_pred = logreg.predict(X_test_vec)

print("Logistic Regression Report:\n", classification_report(y_test, y_pred))
```

**4.Confusion Matrix and Accuracy Plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10,6))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=logreg.classes_, yticklabels=logreg.classes_, cmap='Blues')
plt.title("Confusion Matrix - Logistic Regression")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

```python
from datasets import Dataset
from transformers import BertTokenizer, BertForSequenceClassification, TrainingArguments, Trainer
import pandas as pd # Import pandas here
import numpy as np # Import numpy here for the prediction part
from sklearn.metrics import classification_report # Import classification_report here

df = None # Initialize df to None

try:
    # Load the dataset and rename columns
    df = pd.read_csv("emotion-labels-test.csv")
    # Rename 'text' to 'clean_text' and 'label' to 'labels'
    df.rename(columns={"text": "clean_text", "label": "labels"}, inplace=True)
except FileNotFoundError:
    print("Error: emotion-labels-test.csv not found. Please make sure the file is in the same directory.")
    # Instead of exit(), you might choose to set a flag or continue,
    # depending on how you want to handle the absence of the file.
    # For this fix, we'll rely on the check below.


# Only proceed if df was successfully loaded
if df is not None and not df.empty:
    # Create Dataset from pandas DataFrame, ensuring 'labels' column is included
    dataset = Dataset.from_pandas(df[['clean_text', 'labels']])
    # Encode the 'labels' column as class labels
    dataset = dataset.class_encode_column('labels')
    train_test = dataset.train_test_split(test_size=0.2)

    tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

    def tokenize_function(examples):
        # Tokenize the text data
        return tokenizer(examples["clean_text"], padding="max_length", truncation=True)

    # Apply the tokenization function to the dataset splits
    tokenized = train_test.map(tokenize_function, batched=True)

    # Load the BERT model for sequence classification
    model = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=dataset.features['labels'].num_classes)

    # Define training arguments
    args = TrainingArguments(
        output_dir="./results",
        eval_strategy="epoch",
        per_device_train_batch_size=8,
        per_device_eval_batch_size=8,
        num_train_epochs=2,
        weight_decay=0.01,
        logging_dir='./logs',
        logging_steps=10,
        # The Trainer will automatically look for a column named 'labels'
        # If your label column had a different name, you would specify it here:
        # label_names=['your_label_column_name']
    )

    # Initialize the Trainer
    trainer = Trainer(
        model=model,
        args=args,
        train_dataset=tokenized["train"],
        eval_dataset=tokenized["test"],
    )


    #Train the model
    trainer.train()

    # Predict and report only if training was successful
    preds_output = trainer.predict(tokenized["test"])
    pred_labels = np.argmax(preds_output.predictions, axis=1)
    true_labels = tokenized["test"]["labels"] # Use 'labels' as the column name here

    print("BERT Classification Report:\n", classification_report(true_labels, pred_labels))
else:
    print("Skipping BERT model training due to missing data.")
```

```python
# Predict and evaluate
predictions = trainer.predict(tokenized["test"])
preds = np.argmax(predictions.predictions, axis=1)
labels = predictions.label_ids

label_names = dataset.features['labels'].names  # <-- Fixed line
pred_labels = [label_names[p] for p in preds]
true_labels = [label_names[t] for t in labels]

print("\nClassification Report:\n")
print(classification_report(true_labels, pred_labels))
```
